```python
import numpy as np

# simplified categories
simple_categories = [
    "Employed full time",
    "Employed part time",
    "Student full time",
    "Student part time",
    "Independent contractor",
    "I prefer not to say",
    "Not employed",
    "Retired"
]
#creating a function to standardize the Employment column

def simplify_employment(value):
    if pd.isna(value): #If missing value, return np.nan
        return np.nan
    
    # Convert to lower case to simplify matching
    v = value.lower()
    
    # I prefer not to say
    if "i prefer not to say" in v:
        return "I prefer not to say"
    
    # Retired
    elif "retired" in v:
        return "Retired"
    
    # Student full/part time
    elif "student, full-time" in v:
        return "Student full time"
    elif "student, part-time" in v:
        return "Student part time"
    
    # Employment full/part time
    elif "employed, full-time" in v:
        return "Employed full time"
    elif "employed, part-time" in v:
        return "Employed part time"
    
    # using the 'or' because there are several situations where I want to convert to Independent contractor
    elif "independent contractor" in v or "freelancer" in v or "self-employed" in v:
        return "Independent contractor"
    
    # Not employed
    elif "not employed" in v:
        return "Not employed"
    
    # Catch-all (optional in case I want to make sure my code doesn't miss any different entries)
    else:
        return "Other"
```

```python
# Applying the simplify_employment function on the 'Employment' column, and assigning it to a new column
df['Employment_simplified'] = df['Employment'].apply(simplify_employment)
df['Employment_simplified'].value_counts()  #counting all the values in the new column, to verify my function works
```

```python
# Creating a function to exploade the multi-answer, semicolon-separated columns into a long format without loosing any data:
# Multi-select survey responses are normalized into separate long-format tables using explode(),
# to preserve all selections while avoiding artificial row duplication caused by misaligned list lengths.

# Columns to explode:
# LanguageHaveWorkedWith  
# LanguageWantToWorkWith  
# DatabaseHaveWorkedWith
# DatabaseWantToWorkWith
# PlatformHaveWorkedWith 
# PlatformWantToWorkWith
# WebframeHaveWorkedWith 
# WebframeWantToWorkWith 

def explode_column(df, id_col, target_col):

    # Creating a copy of the original df incase I make mistakes, I won't change the original dataset. 
    # Keep only ID column and the target column 
    temp_df = df[id_cols + [target_col]].copy()

    # Removing rows where the target column is missing
    temp_df = temp_df.dropna(subset=[target_col])

    # Spliting the semicolon-separated string into a list
    temp_df[target_col] = temp_df[target_col].str.split(';')

    # Exploding the list so each value becomes its own row
    temp_df = temp_df.explode(target_col)

    # Removing whitespaces
    temp_df[target_col] = temp_df[target_col].str.strip()

    # Reseting index for cleanliness
    temp_df = temp_df.reset_index(drop=True)

    return temp_df
```

```python
# Applying the function to LanguageHaveWorkedWith column
id_cols = ['ResponseId', 'Country', 'WorkExp','Employment_simplified', 'EdLevel']  #defining these columns to keep track of each respondents Languages and for further exploration

languages_have = explode_column(df, id_cols, 'LanguageHaveWorkedWith')
languages_have
```

```python
###Visualizing Languages worked with by different counties


# Step 1: Grouping the df by the Country column, and finding the top 5 languages worked with per country
lang_per_country = languages_have.groupby('Country')[['LanguageHaveWorkedWith']].value_counts().head(5)
lang_per_country
```

```python
# Step 2: Count how many respondents selected each Language per country

# Group by Country and LanguageHaveWorkedWith, then count the occurrences
counts = languages_have.groupby(['Country', 'LanguageHaveWorkedWith']).size().reset_index(name='count')

counts
```

```python

# Step 3: Find the top 5 databases per age group

# Use groupby + nlargest to keep only the top 5 Languages per country
top5_per_country = counts.groupby('Country').apply(lambda x: x.nlargest(5, 'count')).reset_index(drop=True)

top5_per_country
```

```python
# Step 4: Compute proportions for each language for each country

# First, get total respondents per country
country_totals = df['Country'].value_counts()

# Merge the totals into top5 dataframe to calculate proportions
top5_per_country = top5_per_country.merge(
    country_totals.rename('country_total'),
    left_on='Country',
    right_index=True
)

top5_per_country
```

```python
# Step 5: Creating a df of the top 30 countries by total counts of languages worked with, to use the best sample size
top30 = top5_per_country.nlargest(150, 'country_total')
top30
```

```python
# Compute the proportion of respondents for each language
top30['proportion'] = top30['count'] / top30['country_total']

top30
```

```python

# Step 6: Pivot the dataframe for plotting

# Pivot so that each row = country, 
# each column = prog language,
# and values = proportion/ % of respondents from the country that have worked with the lang
pivot_df = top30.pivot(
    index='Country',
    columns='LanguageHaveWorkedWith',
    values='proportion'
).fillna(0)  # fill NaN with 0 for Languages not in top 5 of that country

pivot_df
```

```python
# Step 6: Plot stacked bar chart
# This is a more inclusive visualization to be able to see what are the most common prog. lang in the most active countries in the survey,
# which allows me to analyze a bigger smaple size to find more conclusive insights. 
pivot_df.plot(kind='bar', stacked=True, figsize=(12,7))

plt.title('Top 5 Languages Respondents Worked with – Top 30 Countries')
plt.ylabel('Proportion of Respondents')
plt.xlabel('Country')
plt.legend(title='Language', bbox_to_anchor=(1.05, 1), loc='upper left')  # place legend outside
plt.tight_layout()
plt.show()
```

```python

#filtering the top30 df to top10 df
top10 = top30.nlargest(50,'country_total')

#Creating the same pivot table from the top10 df 
# Pivot so that each row = country, 
# each column = prog language,
# and values = proportion/ % of respondents from the country that have worked with the lang
top10_pivot = top10.pivot(
    index = 'Country',
    columns = 'LanguageHaveWorkedWith',
    values = 'proportion'
).fillna(0)  # fill NaN with 0 for Languages not in top 5 of that country

top10_pivot
```
